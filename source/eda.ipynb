# EDA

Vamos a analizar cualitativamente el dataset en base a la información provista en el README.

Tenemos 3 archivos de formato .dat: MOVIES, RATINGS y USERS.

Estos archivos contienen **1,000,209** valoraciones anónimas de aproximadamente **3,900** películas hechas por **6,040** usuarios de MovieLens que se unieron a MovieLens en el año 2000.

RATINGS
================================================================================

Todas las valoraciones están contenidas en el archivo "ratings.dat" y están en el
siguiente formato:

UserID::MovieID::Rating::Timestamp

- Los UserIDs van desde 1 hasta 6040
- Los MovieIDs van desde 1 hasta 3952
- Las valoraciones se realizan en una escala de 5 estrellas (solo valoraciones de estrellas enteras)
- El Timestamp está representado en segundos desde el epoch, como lo devuelve time(2)
- Cada usuario tiene al menos 20 valoraciones

USERS
================================================================================

La información de los usuarios está en el archivo "users.dat" y tiene el siguiente
formato:

UserID::Gender::Age::Occupation::Zip-code

Toda la información demográfica es proporcionada voluntariamente por los usuarios y no se
verifica su precisión. Solo se incluyen en este conjunto de datos los usuarios que han proporcionado
alguna información demográfica.

- El género está representado por una "M" para masculino y "F" para femenino
- La edad se elige de los siguientes rangos:

	*  1:  "Menor de 18"
	* 18:  "18-24"
	* 25:  "25-34"
	* 35:  "35-44"
	* 45:  "45-49"
	* 50:  "50-55"
	* 56:  "56+"

- La ocupación se elige de las siguientes opciones:

	*  0:  "otro" o no especificado
	*  1:  "académico/educador"
	*  2:  "artista"
	*  3:  "administrativo/clerical"
	*  4:  "estudiante universitario/postgrado"
	*  5:  "atención al cliente"
	*  6:  "doctor/profesional de salud"
	*  7:  "ejecutivo/gerencial"
	*  8:  "agricultor"
	*  9:  "ama de casa"
	* 10:  "estudiante de primaria/secundaria"
	* 11:  "abogado"
	* 12:  "programador"
	* 13:  "jubilado"
	* 14:  "ventas/marketing"
	* 15:  "científico"
	* 16:  "autónomo"
	* 17:  "técnico/ingeniero"
	* 18:  "artesano"
	* 19:  "desempleado"
	* 20:  "escritor"

MOVIES
================================================================================

La información de las películas está en el archivo "movies.dat" y tiene el siguiente
formato:

MovieID::Title::Genres

- Los títulos son idénticos a los títulos proporcionados por IMDB (incluyendo
el año de estreno)
- Los géneros están separados por barras verticales (|) y se seleccionan de los siguientes géneros:

	* Acción
	* Aventura
	* Animación
	* Infantil
	* Comedia
	* Crimen
	* Documental
	* Drama
	* Fantasía
	* Cine negro
	* Terror
	* Musical
	* Misterio
	* Romance
	* Ciencia ficción
	* Suspenso
	* Guerra
	* Western

- Algunos MovieIDs no corresponden a una película debido a entradas duplicadas accidentales
y/o entradas de prueba
- Las películas son mayormente ingresadas a mano, por lo que pueden existir errores e inconsistencias
